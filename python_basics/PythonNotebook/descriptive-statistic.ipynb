# Descriptive statistics

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('..\Data\percent_bachelors_degrees_women_usa.csv')

df.head()
```

```python
data = [100 , 20 ,5, 20 , 45 , -100 , 46 , 50]
```

# Mean 
 - sum(n)/count(n)

```python
mean =  np.mean(data) # mean is equal to avg
round(mean,1)
```

# Median
  - mid element after sorting

```python
data.sort()
data
```

```python
median = np.median(data)
round(median,1)
```

# Mode
    - value repeating the most

```python
# mode is not available in numpy use scipy
from scipy import stats
```

```python
mode = stats.mode(data)

mode
```

# Variance and standard deviation
    - how far a number is from mean
    - std square root of variance

```python
variance = np.var(data)

variance
```

```python
std = np.std(data)

std
```
