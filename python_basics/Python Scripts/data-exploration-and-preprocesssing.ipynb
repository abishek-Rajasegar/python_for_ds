# Agg , Filtering Sorting Grouping

```python
data_df = pd.DataFrame({
    'name': [ 'Ram' , 'Sam' , 'Arun' , 'Ajay' ],
    'age': [22,21,21,22],
    'salary':[100,200,90,150],
    'Dept':['IT','Acc','Hr','Ma']
})

data_df
```

```python
data_df = data_df.sort_values(by="name", ascending=False )
# sort desc
data_df
```

```python
# grouping
# all data
data_df.groupby("age").count()
```

```python
# specific column group by
data_df.groupby("age")["name"].count()
```

```python
# Avg salary based on age
data_df.groupby("age")["salary"].mean()
```
